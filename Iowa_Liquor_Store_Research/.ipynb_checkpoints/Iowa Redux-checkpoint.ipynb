# Stage 1: Data Massaging
#### 1a. read in the data locally.

```python
import pandas as pd
iowa = pd.read_csv('Data/Iowa_Liquor_Sales_reduced.csv',low_memory=False)

iowa.head(3)
```

**First things first, we are going to have to make some changes to our data.**
- Convert 'Date' to a Datatime format.
- Remove dollar signs from all dollar figures to convert to float.
- Remove unnecessary features
> - Zip Code, ( I think 'City' will suffice as a feature and to split zip into dummies would create a lot of features)
> - Category (Number),Item Number & Description, Vendor Number
> - Volume Sold (Gallons)''
- Convert 'State Bottle Cost' and 'State Bottle Retail' to 'Bottle Markup'.  This will be averaged once observations are aggregated by store
- Use Mode Bottle Volume(ml) per store.  

#### 1b. remove unnecessary columns

```python
# Dropping columns that will not be used.
iowa.drop(['Zip Code','County Number','Category','Vendor Number','Item Number','Item Description','Volume Sold (Gallons)'], axis =1, inplace = True)
iowa.head(3)
```

#### 1c. convert date to python datetime data type

```python
# Convert the Data column to a datetime dtype.
iowa["Date"] = pd.to_datetime(iowa["Date"], format="%m/%d/%Y")
```

#### 1d. removing string elements and converting data types.

```python
# Removing the dollar sign from sales values so I can convert to a float.
iowa['Sale (Dollars)'] = iowa['Sale (Dollars)'].map(lambda x: x.lstrip('$'))
iowa['State Bottle Cost'] = iowa['State Bottle Cost'].map(lambda x: x.lstrip('$'))
iowa['State Bottle Retail'] = iowa['State Bottle Retail'].map(lambda x: x.lstrip('$'))

# Converting to a float
iowa[['Sale (Dollars)','State Bottle Cost','State Bottle Retail']] = iowa[['Sale (Dollars)','State Bottle Cost','State Bottle Retail']].apply(pd.to_numeric)
```

```python
# Converting store number to a string dtype
iowa['Store Number'] = iowa['Store Number'].astype(str)
```

#### 1e. engineering a 'state markup' feature.

```python
# Engineering a feature of State Markup.  
iowa['State Markup'] = ((iowa['State Bottle Retail']/iowa['State Bottle Cost']) - 1).round(decimals = 3)

iowa.head()
```

#### 1f. splitting data by year

```python
# Splitting data by year.
iowa2016 = iowa[iowa['Date'].dt.year == 2016]

iowa2015 = iowa[iowa['Date'].dt.year == 2015]

```

This long list of Cities who are missing county values is no bueno.  I will have to make a dictionary of City and Country names to use to fill these values.  I can't imagine that Iowa would just go and add 20+ cities one year and I would also like to assume that those cities had corresponding counties the previous year which I can use for this key value pair filling.

County is going to be my dictionary key and a list of cities names will be the values.

I noticed that there are some data entry errors where apostrophies are used.  
"ARNOLD'S PARK:Dickinson" &'ARNOLDS PARK:Dickinson' both appeared in the City-County Set, so I will be removing apostrophes.

```python
# getting city/county pairs into a set
iowa_cities = set(iowa2015['City'] +':'+ iowa2015['County'])
# result = 'City:County'

# Splitting city-county strings into two parts.
new_iowa = []
for sett in iowa_cities:
    sett = sett.replace("'", "")
    new_iowa.append(sett.split(':'))    
#Result = ['City','State]


# Converts the column to a dictionary where keys are Counties and values are lists of cities.
iowa_dict = dict()
for combo in new_iowa:
    if combo[1] in iowa_dict.keys():
        iowa_dict[combo[1]].append(combo[0])
    else:
        iowa_dict[combo[1]] = [combo[0]]
#Result = 'County' : ['City1','City2','City3',]
```

```python
# Just taking a look at the City/County Dictionary.
for key, value in iowa_dict.iteritems():
    print key, value
```

#### 3b. create and apply function to generalize category based on key words.

```python
# FUNCTION FOR SIMPLIFYING THE CATEGORY TYPES
def simple_alchy(category):
    if 'VODKA' in category:
        return 'VODKA'
    elif 'RUM' in category:
        return 'RUM'
    elif 'SCHNAPPS' in category:
        return 'SCHNAPPS'
    elif 'WHISKIES' in category:
        return 'WHISKIES'
    elif 'WHISKEY' in category:
        return 'WHISKIES'
    elif 'TEQUILA' in category:
        return 'TEQUILA'
    elif 'GINS' in category:
        return 'GIN'
    elif 'SCOTCH' in category:
        return 'SCOTCH'
    elif 'BRANDIES' in category:
        return 'BRANDY'
    elif 'CORDIALS' in category:
        return 'CORDIALS'
    elif 'LIQUEURS' in category:
        return 'LIQUEURS'
    else:
        return 'OTHER'
    
```

```python
#Applying the alchy function
iowa2015['Category Name'] = iowa2015['Category Name'].apply(lambda x: simple_alchy(x)) 

iowa2015['Category Name'].value_counts()
```

### Flattening the Data by making columns for every monthly value

    - AVG State Bottle Retail(JAN).->...AVG State Bottle Retail(JAN)

```python
# creating this dict to map string months to numerical months for the purpose of simplifying the flattening process.
month_dict= {1:'JAN', 2:'FEB', 3:'MAR', 4:'APR', 5:'MAY', 6:'JUN',7:'JUL',8:'AUG',9:'SEP',10:'OCT',11:'NOV',12:'DEC'}

# for every store number in Iowa
for store_num in stores:
    # start by creating a dataframe slice of just the stores information.
    for index, row in cat_store_sales[cat_store_sales['Store Number'] == store_num].iterrows():
        # if the store has information for january than we will.
        for item in months:
            if row['month'] == item:
                MON = month_dict[item]
                mon_feat_df.set_value(store_num, 'Sale (Dollars)('+MON+')', row['Sale (Dollars)'])
                mon_feat_df.set_value(store_num, 'Volume Sold (Liters)('+MON+')', row['Volume Sold (Liters)'])
                mon_feat_df.set_value(store_num, 'Bottles Sold('+MON+')', row['Bottles Sold'])
                mon_feat_df.set_value(store_num, 'AVG Bottle Vol (ml)('+MON+')', row['AVG Bottle Vol (ml)'])
                mon_feat_df.set_value(store_num, 'AVG State Bottle Cost('+MON+')', row['AVG State Bottle Cost'])
                mon_feat_df.set_value(store_num, 'AVG State Bottle Retail('+MON+')', row['AVG State Bottle Retail'])
                mon_feat_df.set_value(store_num, 'AVG State Markup('+MON+')', row['AVG State Markup'])
        # worked perfectly first try.  
```

Data that is going to be passed as training data must meet the below criteria.
1. Must have the aggregate and mean sales features for JAN, FEB & MAR; These features are available for the 2016 data and are the only months of overlap we have.

Mode Bottle Volume:  alcohol has 6 basic sizes as far as volume, typically with associations about each size:
50ml (airplane), 375ml, 500ml, 750ml, 1L and 1.5L

375ml and 500ml can (sadly) be associated with alcoholism.  Individuals will by what they can consume in a single evening as their alcoholic tendencies cause them to want to drink until all is gone.

50ml and 1.5L for the people that need to drink all the time everywhere.  College kids and recent college grads.

These are generalized opinions that do not have definative research backings behind them.  The point was not to illustrate the these are hard and fast truths, but specific demographics can be more likely to purchase a specific bottle size .If we averaged a locations where 1.5L bottles and 500ml bottles are most commonly sold, our mean will fall around 1L wrongly Indicating 1 demographic where two could possibly exist.
