```python
import pandas as pd
iowa2015 = pd.read_csv('Data/iowa2015_model_data.csv')
iowa2016 = pd.read_csv('Data/iowa2016_model_data.csv')
```

```python
#viewing data
#change view options of dataframe
pd.options.display.max_columns = 100

# removing stores without full values. because annual sales are a calculation of monthly total sales, 
# store missing months are going to have annual sales lower than they should be.
iowa2015.dropna(inplace=True)
iowa2015.head()
```

```python
# taking out just the features from the first 3 months as these are our predictive values.
fit_features = list(iowa2015)

fit_features = fit_features[0:23]+fit_features[-2:]

iowa2015_fit= iowa2015[fit_features]

```

Before this data can be modeled with I am going to need to do a little pre-processing.
-  Categorical Features, County & City, will need to be converted to dummy variables.
-  Numerical features will need to be scaled to improve models predicatbility. 
     - AVG State Markup already exists on a standard/normal scale so this will be left alone.  
     - Im choosing a standardization technique as it looks like alot of annual sales fall in the same region.  There is a relatively normal distribution with a right skew from a select handful of extremely well performing stores.  The standardization will help with allowing outlier information to help with the identification of these extremely high performing scenarios.

```python
# quick view of the distribution
import seaborn as sns
% matplotlib inline
sns.distplot(iowa2015_fit['Annual Sale(Dollars)'])
```

#### Getting dummie values 

```python
iowa2015_dummies = iowa2015[['Store Number','City','County']]
iowa2015_dummies.set_index(keys='Store Number', inplace=True)
iowa2015_dummies = pd.get_dummies(iowa2015_dummies, prefix =['city','county'],drop_first = True)
```

#### Standardizing other values.

```python
# fit_features = list(iowa2015)
# fit_features = fit_features[0:23]+fit_features[-2:]

# iowa2015_scale = iowa2015[fit_features]
# iowa2015_scale.set_index(keys='Store Number',inplace =True)
# iowa2015_scale.drop(['Annual Sale(Dollars)','City','County'],axis=1,inplace=True)



iowa2015_scale.head()
```

```python
'''Scaling returns a numpy matrix without indexes or feature labels, 
so I will need to specify those values when converting the matrix to a dataframe to merge with the dummied info.'''

# index values for scaled dataframe
iowa2015_stores = iowa2015_scale.index.values
# feature headers
iowa2015_scale_features = list(iowa2015_scale)

# import scaler
from sklearn.preprocessing import StandardScaler
# set scaler
SS = StandardScaler()
# fit scaler
iowa2015_scaled = pd.DataFrame(data = SS.fit_transform(iowa2015_scale),index =iowa2015_stores, columns = iowa2015_scale_features)

iowa2015_scaled.head()
```

```python
# removing the index name for easier merge.
del iowa2015_dummies.index.name

creating a 3rd dataframe of annual sales for merging
iowa2015_ann_sales = iowa2015_fit[['Store Number','Annual Sale(Dollars)']]
iowa2015_ann_sales.set_index(keys='Store Number',inplace=True)
del iowa2015_ann_sales.index.name


# mergine Scaled, Dummies and Sales information into a final DF.

iowa2015_model = pd.concat([iowa2015_ann_sales, iowa2015_scaled, iowa2015_dummies], axis=1)
iowa2015_model.head()
```

```python
# creating X and y variables
X = iowa2015_model.drop('Annual Sale(Dollars)',axis =1)
y = iowa2015_model['Annual Sale(Dollars)']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)


from sklearn.linear_model import LinearRegression

LR = LinearRegression(n_jobs = -1)


LR.fit(X_train, y_train)
Y_pred = LR.predict(X_test)

LR.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, Y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, Y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, Y_pred)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(y_test, Y_pred)
plt.xlabel('Y True')
plt.ylabel('Y Pred')
plt.show()
```

#### It seems that using the standardized data produce results was no bueno just absolutely aweful...  

Lets go back and use the non standardized data.  Some of the features were monthly sales which hold a direct relation to the variable we are predicting and so standardizing them could cause some issues.

```python
del iowa2015_scale.index.name

iowa2015_model2 = pd.concat([iowa2015_ann_sales,iowa2015_scale, iowa2015_dummies,],axis=1)
```

```python
# creating X and y variables
X2 = iowa2015_model2.drop('Annual Sale(Dollars)',axis =1)
y2 = iowa2015_model2['Annual Sale(Dollars)']

from sklearn.model_selection import train_test_split

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2,y2, test_size=0.25)


from sklearn.linear_model import LinearRegression

LR2 = LinearRegression(n_jobs = -1)


LR2.fit(X_train2, y_train2)
y_pred2 = LR2.predict(X_test2)

LR2.score(X_test2,y_test2)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test2, y_pred2)
# that is a nice R2
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test2, y_pred2)
```

On average our predictions are off by almost $40000, this has the potential to be significant if you think about stores whose annual sales are in the 100K to 300K range, however large stores that see 800K+ sales annually dont see this as such a large swing.  

```python
# probably not the best metric considering we're measuring sales where an error of 3000 dollars is not significant
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test2, y_pred2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(y_test2, y_pred2)
plt.xlabel('Y True')
plt.ylabel('Y Pred')
plt.show()
```
